```python
import pandas as pd
import numpy as np

# reading data from dataset in a pandas df
diabetes_data = pd.read_csv("diabetes.csv")
```

```python
# what are the columns in this df
diabetes_data.columns
```

```python
# number of rows/records in dataframe
len(diabetes_data)
```

```python
# lets find the total number of null values in the different columns
diabetes_data.isnull().sum()
```

```python
# let's look at the data a little more closely
# first some basic statistics
diabetes_data.describe()
```

```python
# looking at the min values for a couple of columns, we know there is a problem with this dataset
# the glucose, BP, skin thickness, insulin and BMI of a person (here a record) cannot be zero
# we must replace these 0 values with pythonic NaN
diabetes_data[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']] = diabetes_data[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']].replace(0,np.NaN)
```

```python
# now lets look at both the sum of null values
diabetes_data.isnull().sum()
```

```python
# as well new basic statistics 
diabetes_data.describe()
```

```python
# now lets look at all the rows/records that contain these null values
diabetes_data[diabetes_data.isna().any(axis=1)]
```

```python
# looks like it's mostly skin thickness and insulin columns
# let's look at column datatypes
diabetes_data.dtypes
```

```python
# exploring values in columns
diabetes_data['Pregnancies'].value_counts()
```

```python
diabetes_data['Age'].value_counts()
```

```python
# replacing BMI null values with mean BMI
mean_BMI = diabetes_data['BMI'].mean()
diabetes_data['BMI'].fillna(value=mean_BMI, inplace=True)
```

```python
# BMI has no null values
diabetes_data.isnull().sum()
```
